```python
# Task 1: Data Retreiving

# import packages and set aliases
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#assign file name to an object
concrete_data_filename = 'Concrete_Data.csv'

# read in data file
concrete = pd.read_csv(concrete_data_filename, sep = ',', decimal = '.', header = None, skiprows = 1, names = ['cement', 'slag','flyash','water','super','coarseagg','fineagg','age','compressive'])

# checking whether the loaded data is equivalent to source file by comparing the head and tail to for obvious mislabelled columns -etc-.
print(concrete.head(10))

print(concrete.tail(10))

# checking if correct column names are entered
print(concrete.columns)

# checking the number of entries and number of attributes and compare them to information given in concrete_readme file
print(concrete.shape)

# checking if python type each column meaningfully
print(concrete.dtypes)

# checking range of values and impossible values
print(concrete.describe())

# checking for number of missing values for each variables
print(concrete.info())

# plotting histogram of compressive strength
concrete['compressive'].plot(kind = 'hist', bins = 20)
plt.title('Histogram of Concrete Compressive Strength (MPa)')
plt.xlabel('Compressive Strength (MPa)')
plt.show()

# recoding numerical variable 'compressive strength' to categorical variable, for less than 40 MPa as non-compliant and more than or equal to 40 MPa as compliant
def compliance(series):
    if series < 40:
        return "non-compliant"
    elif series >= 40:
        return "compliant"

concrete['compressive'] = concrete['compressive'].apply(compliance)

# subset the dataset to only include those rows whose age is 28 days
concrete = concrete[concrete['age'] == 28]

# dropping age variable
concrete.drop('age', axis = 1, inplace = True)

# checking if dropping was successful
concrete.shape


# checking counts of each level of newly recoded categorical variable 'compressive'
concrete['compressive'].value_counts(sort=False)
```

```python
# Task 3 Data Modelling

# recoding compressive strength to 0 and 1 for modelling purposes
def targeting(series):
    if series == "non-compliant":
        return 0
    elif series == "compliant":
        return 1

concrete['compressive'] = concrete['compressive'].apply(targeting)

# checking if conversion worked
concrete.compressive.head()

# setting variable compressive as dependent target
target = concrete[['compressive']]

# setting the rest of the variables as features
features = concrete.drop("compressive", axis = 1)

# standardising 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

features = scaler.fit_transform(features)
#import train test split
from sklearn.model_selection import train_test_split

# import packages for modelling
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import make_scorer, recall_score, precision_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
# K-nearest neighbor 50% for training and 50% for testing 
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size = 0.5, random_state = 42, stratify = target)


#Fitting basic model
#initialise k-nearest neighbors model
model = KNeighborsClassifier()

#fitting model on temporary data set
model.fit(features_train, target_train)

print("Basic Model")
print("Model Accuracy on Training Set: % 3.5f percents" %(model.score(features_train, target_train)*100))# accuracy of the model on training set
print("Model Accuracy on Testing Set: % 3.5f percents" %(model.score(features_test, target_test)*100)) # accuracy of th emodel on test set

# using the model to predict target based on features in testing set
target_test_predict = model.predict(features_test)

# import confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(target_test, target_test_predict)
# print confusion matrix
print cm

# import and print modelling performance
from sklearn.metrics import classification_report
print classification_report(target_test, target_test_predict)

# using precision as scoring metric
precision = make_scorer(precision_score)





# perform 10 fold cross validation 
from sklearn.model_selection import ShuffleSplit
cv = ShuffleSplit(n_splits = 10, test_size = 0.5, random_state = 42)
cv_results = cross_val_score(estimator = model, X = features, y = target, cv = cv, scoring = precision)

# print cross validation result
print(cv_results)

# print the mean and standard deviation of precision score on all 10 fold results
print('The mean: {}'.format(cv_results.mean()))
print('The std: {}'.format(cv_results.std()))

# smaller standard deviation, the tighter the ten means. This indicates the actual accuracy
# of new data will probably match the mean of the cross validation score pretty well. 





##Start parameter tuning
# print model's hyperparameters
print(model.get_params())

## setting range of values of hyperparameters for tuning
n_neighbors = [i for i in range(3, 31, 1)] # set from 3 to 30 increment of 1
weights = ["uniform"] # choose uniform for weight parameter, meaning all points in each neighborhood are weighted equally
p = [1, 2] # p = 1 to use manhattan distance, to make other features playing role in the distance calculation in our multidimensional space. 

# combine all hyperparameters into a dictionary
parameters = dict(n_neighbors = n_neighbors, weights = weights, p = p)

#printinting dictionary of hyperparameters
parameters

# initialise grid search algorithm 
param_search = GridSearchCV(model, parameters, cv = 10, scoring = precision)

# use grid search algorithm to fit the training set to search for best combination of hyperparameters
param_search.fit(features_train, target_train)

# print best parameters to achieve highest precision score
print(param_search.best_params_)

#print best estimator resulting from grid search algorithm
print(param_search.best_estimator_)

# assign grid search cv best estimator result as a new model
model_val = param_search.best_estimator_

#fitting best estimator model on training set
model_val.fit(features_train, target_train)

#use new model to predict validation set
prediction_test = model_val.predict(features_test)

#accuracy on training set
print("Model after tuning")
print("Model After Tuning Accuracy on Training Set: % 3.5f percents" %(model_val.score(features_train, target_train)*100)) 
print("Model After Tuning Accuracy on Testing Set: % 3.5f percents" %(model_val.score(features_test, target_test)*100))
cm = confusion_matrix(target_test, prediction_test)
print cm
print classification_report(target_test, prediction_test)   




##perform 10 fold cross validation on validated model
from sklearn.model_selection import ShuffleSplit
cv = ShuffleSplit(n_splits = 10, test_size = 0.5, random_state = 42)
cv_results = cross_val_score(model_val, features, target, cv = cv, scoring = precision)

#print 10 fold cross validation results
print("10-fold Cross Validation Results of Model After Tuning")
print(cv_results)

## print the mean and standard deviation of all precision scores from 10 folds. 
print('The mean: {}'.format(cv_results.mean()))
print('The std: {}'.format(cv_results.std()))

# smaller standard deviation, the tighter the 10 means are. This indicates the actual accuracy
# of new data will probably match the mean of the cross validation score fairlywell. 
```
